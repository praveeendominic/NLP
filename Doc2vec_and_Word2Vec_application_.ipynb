```python
def tagged_document(list_of_list_of_words):
   for i, list_of_words in enumerate(list_of_list_of_words):
      yield gensim.models.doc2vec.TaggedDocument(list_of_words, [i])
```

```python
data=list(tagged_document(reviews))
data
```

```python
from gensim.models import Doc2Vec
```

```python
model_dv= Doc2Vec()
```

```python
model_dv.build_vocab(data)
```

```python
model_dv.train(data, total_examples=model_dv.corpus_count, epochs=model_dv.epochs)
```
