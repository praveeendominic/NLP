```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Practice/Fake news classifier/test.csv')
df1.head()
```

```python
df2=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Practice/Fake news classifier/submit.csv')
df2.head()
```

```python
df1.shape,df2.shape
```

```python
df=df1.merge(df2,
          how="inner",
        on='id')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
sentences=list(df.title)
sentences
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('all')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
corp=[]
```

```python
lemmatizer=WordNetLemmatizer()
for i in range(len(sentences)):
  sent=re.sub('[^a-zA-Z]', ' ',sentences[i]).lower()
  words= nltk.word_tokenize(sent)
  words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
  corp.append(" ".join(words))
corp
```

```python
corp[:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
X = cv.fit_transform(corp).toarray()
```

```python
X.shape
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,random_state=3)
```

```python
X_train.shape,X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()

nb.fit(X_train, y_train)
```

```python
y_pred=nb.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
```
