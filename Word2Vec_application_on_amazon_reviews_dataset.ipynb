```python
df=pd.read_json('/content/drive/MyDrive/Colab Notebooks/NLP/Word2vec/Cell_Phones_and_Accessories_5.json',lines=True)
df.head()
```

```python
import gensim
gensim.utils.simple_preprocess("They look good and stick good! I just don't.")
```

```python
reviews=df.reviewText.apply(gensim.utils.simple_preprocess)
reviews
```

```python
from gensim.models import Word2Vec
```

```python
model=Word2Vec(window=3, min_count=2)
model.build_vocab(reviews)
```

```python
model.most_similar('great')
```

```python
model.similarity('awsome','fantastic')
```
